```python
import numpy as np
import matplotlib.pyplot as plt

# 設定極座標和花瓣數量
theta = np.linspace(0, 2*np.pi, 400)
k = 5
r = np.sin(k * theta)

# 繪製極座標圖
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')  # 設定極座標
ax.plot(theta, r, color='pink', linewidth=2)

# 美化圖形(移除極座標刻度,移除半徑標籤,移除角度標籤)
ax.set_rticks([])
ax.set_yticklabels([])
ax.set_xticklabels([])

ax.set_title("Flower", fontsize=18, color='blue')

plt.show()
```
