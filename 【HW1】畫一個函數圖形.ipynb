# üå∏ Ê•µÂ∫ßÊ®ôËä±Áì£Êõ≤Á∑ö (Rose Curve) üé®

Áπ™Ë£Ω **Ê•µÂ∫ßÊ®ôËä±Áì£Êõ≤Á∑öÔºàRose CurveÔºâ**  
ÂÆÉÁöÑÊï∏Â≠∏Ë°®ÈÅîÂºèÁÇ∫Ôºö
\[
r = sin(k*theta)
\]
ÂÖ∂‰∏≠Ôºö
- *\( k \) Ê±∫ÂÆöËä±Áì£Êï∏Èáè*
- Âú®Ê•µÂ∫ßÊ®ô‰∏äÁπ™Âá∫Â∞çÁ®±ÁöÑÂúñÊ°àÔºå‰∏¶Â∞áÂÖ∂Êõ≤Á∑öË®≠ÂÆöÁÇ∫ **Á≤âÁ¥ÖËâ≤**Ôºå‰∏¶ÁßªÈô§‰∫ÜÊâÄÊúâËª∏Ê®ôÁ±§ÔºåËÆìÁï´Èù¢Á∞°ÊΩî„ÄÇüå∑


```python
import numpy as np
import matplotlib.pyplot as plt

# Ë®≠ÂÆöÊ•µÂ∫ßÊ®ôÂíåËä±Áì£Êï∏Èáè
theta = np.linspace(0, 2*np.pi, 400)
k = 5
r = np.sin(k * theta)

# Áπ™Ë£ΩÊ•µÂ∫ßÊ®ôÂúñ
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, r, color='pink', linewidth=2)

# ÁæéÂåñÂúñÂΩ¢(ÁßªÈô§Ê•µÂ∫ßÊ®ôÂàªÂ∫¶,ÁßªÈô§ÂçäÂæëÊ®ôÁ±§,ÁßªÈô§ËßíÂ∫¶Ê®ôÁ±§)
ax.set_rticks([])
ax.set_yticklabels([])
ax.set_xticklabels([])

# Ë®≠ÂÆöÊ®ôÈ°å
ax.set_title("Rose Curve", fontsize=18, color='blue')

plt.show()
```
